### Strings - Anything that is present between the quotation marks - ""

```python
print("hello")
```

```python
# We can also take specific characters from a string - counting starts from 0
print("hello"[0])
```

```python
# We can also extract characters using negative numbers - negative numbers will count from the end.
print("hello"[-1])
```

```python
# Strings can also be concatenated
"123" + "456"
```

### Type Conversion - Changing the datatype

```python
print(int("123") + int("456"))
```

We can convert values to int, str, float, bool

```python
# Running a code with type conversion - The reason we convert this to a string - because we cannot concatenate a string to an int.
print("Number of letters in your name: " + str(len(input("Enter your name: "))))
```

```python
userinput_name = input("Enter your name: ")
len_name = len(userinput_name)

print(f"Number of letters in your name: " , len_name)
```
