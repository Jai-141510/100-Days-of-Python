### Making a program that will calculate the cost of the pizza based on user inputs

```python
print("Welcome to Python Pizza Deliveries!")

userinput_size = input("What size of pizza would you like? S, M or L? ")
userinput_pepperoni = input("Would you like pepperoni on your pizza? Type y for yes and n for no: ")
userinput_cheese = input("Would you like extra cheese on your pizza? Type y for yes and n for no: ")

if userinput_size == "S":
    cost = 15
elif userinput_size == "M":
    cost = 20
else:
    cost = 25

if userinput_pepperoni == "y":
    cost += 2 if userinput_size == "S" else 3

if userinput_cheese == "y":
    cost += 1

print(f"The cost of your pizza will be: ${cost}")
```
