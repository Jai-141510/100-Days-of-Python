## Python Strings

```python
print("Hello world")
```

```python
# Printing multiple lines using 1 print function - backslash n -> \n
print("Hello world! \nHello world 2!")
```

```python
# String concatenation
print("Hello" + " User")
```

```python
# Giving user the ability to enter values -> input()
input("What is your name?")
```

```python
# Printing out the name of the user using input and print together
print("Hello " + input("What is your name?") + "!")
```

## Python Variables

```python
# Assigning variable to the input function
name = input("What is your name?")
print(name)
```

```python
# Trying to find out length of the name the user inputs
name = input("What is your name?")
length = str(len(name))
print("Hi "+ name + "\nYour name has " + length + " characters")
```
