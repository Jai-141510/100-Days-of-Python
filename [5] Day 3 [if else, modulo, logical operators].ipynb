### If else conditional statements

```python
# Writing a very basic if else statement for understanding purposes
userinput_height = int(input("What is your height in cms: "))

if userinput_height > 120:
    print("You can go ahead with the ride")
else:
    print("Sorry, your height is too short for your ride safety")
```

### Modulo operator and it's use case

Modulo operator = % - gives out the reminder of a division. Eg: 10 % 3 will be 1 as 1 is the reminder when you divide it.

```python
# Modulo use case - Checking if the number given by the user is odd or even

userinput_number = int(input("Please enter a random number: "))

if userinput_number % 2 == 0:
    print(f"Your number {userinput_number} is even")
else:
    print(f"Your number {userinput_number} is odd")
```

### Nested if else statement

```python
# Using the previous if else statement and building a nest

userinput_height = int(input("What is your height in cms: "))

# Learning - the outer block runs first, and then it enters the inner nested block if the outer block condition is satisfied
if userinput_height > 120:
    if int(input("What is your age?: ")) > 18:
        print("Please pay $12 to enter")
    else:
        print("Please pay $7 to enter")
else:
    print("Sorry, your height is too short for your ride safety")
```

### Using elif for multiple if statements

```python
# Using the previous if else statement and inputting elif conditions

userinput_height = int(input("What is your height in cms: "))

# Learning - the outer block runs first, and then it enters the inner nested block if the outer block condition is satisfied
if userinput_height > 120:
    userinput_age = int(input("What is your age?: "))
    if userinput_age <= 12:
        print("Please pay $5 to enter")
    elif userinput_age <= 18:
        print("Please pay $7 to enter")
    else:
        print("Please pay $12 to enter")
else:
    print("Sorry, your height is too short for your ride safety")
```

### Multiple if statements

```python
# Using the previous if else statement and inputting multiple if conditions

userinput_height = int(input("What is your height in cms: "))

# Learning - the outer block runs first, and then it enters the inner nested block if the outer block condition is satisfied
if userinput_height > 120:
    userinput_age = int(input("What is your age?: "))
    if userinput_age <= 12:
        bill = 5
        print("Children tickets are $5 to enter")
    elif userinput_age <= 18:
        bill = 7
        print("Youth tickets are $7 to enter")
    else:
        bill = 12
        print("Adult tickets are $12 to enter")
    # The next block will carry the prices of the photos - if the user wants photos to be taken - standard price of $3 added to the bill
    userinput_photos = input("Do you want photos to be taken? Photos are at standard price of $3. Press y for yes and n for no: ")
    if userinput_photos == "y":
        bill += 3
        print(f"Your total bill will be: {bill} which includes $3 for photos")
    else:
        print(f"Your total bill will be: {bill}")
else:
    print("Sorry, your height is too short for your ride safety")
```

### Logical operators

Logical operators allows us to combine multiple conditions in a single line of code

and or not - these 3 are the logical operators that we will be using

```python
# Using the previous if else statement and inputting logical operators

# We will not charge anything for people who are between 45-55 years old as they might be going through midlife crisis

userinput_height = int(input("What is your height in cms: "))

# Learning - the outer block runs first, and then it enters the inner nested block if the outer block condition is satisfied
if userinput_height > 120:
    userinput_age = int(input("What is your age?: "))
    if userinput_age <= 12:
        bill = 5
        print("Children tickets are $5 to enter")
    elif userinput_age <= 18:
        bill = 7
        print("Youth tickets are $7 to enter")
    elif userinput_age >= 45 and userinput_age <= 55: # The use of logical operators
        bill = 0
    else:
        bill = 12
        print("Adult tickets are $12 to enter")
    # The next block will carry the prices of the photos - if the user wants photos to be taken - standard price of $3 added to the bill
    userinput_photos = input("Do you want photos to be taken? Photos are at standard price of $3. Press y for yes and n for no: ")
    if userinput_photos == "y":
        bill += 3
        print(f"Your total bill will be: {bill} which includes $3 for photos")
    else:
        print(f"Your total bill will be: {bill}")
else:
    print("Sorry, your height is too short for your ride safety")
```
